### Example 1: Irreducible S^2

```python
X = Poset({'w': ['x', 'z'], 'y': ['x', 'z'], 'x': ['t', 'b'], 'z': ['t', 'b']})

X.show()
```

There is no acyclic matching $\Sigma$ such that $Flow_\Sigma[x,y]$ is contractible for all critical cells $x,y$

```python
regular_acyclic_matching(X)
```

### Example 2: Reducible S^2

```python
K = simplicial_complexes.Sphere(2)
X = K.face_poset()
X.show()
```

```python
M = regular_acyclic_matching(X)
print(M) 
```

```python
critical_poset(X,M)
```

```python
localization_poset(X,M)
```

### Example 3: Projective Plane

```python
K = simplicial_complexes.ProjectivePlane()
X = K.face_poset()
len(X)
```

```python
X.show()
```

```python
M = regular_acyclic_matching(X)
len(M)
```

```python
critical_poset(X,M)
```

This is the minimal finite model of the projective plane

### Torus

```python
K = simplicial_complexes.Torus()
X = K.face_poset()
len(X)
```

```python
M = regular_acyclic_matching(X)
len(M)
```

```python
C = critical_poset(X,M)
C.show()
```

This is the minimal finite model of the torus.
