### Example 1

```python
X = Poset({'w': ['x', 'z'], 'y': ['x', 'z'], 'x': ['t', 'b'], 'z': ['t', 'b']})
#This is the face poset of the regular CW-structure of a here with two 0-cells 'y', 'w', two 1-cells 'x', 'z', and two 2-cells 't', 'b'

X.show()
```

```python
# Compute the entrance path poset for the interval [w, t]
entrance_poset = entrance_path_poset(X, 'w', 't')
entrance_poset.plot()
```

```python
#Define a matching and compute the Flow category
M = [('y','x'), ('z','b')]
F = flow_category(X, M)
```

```python
#Compute the Flow poset for ('w', 't')
F[('w','t')].plot(figsize=[6,6])
```

```python
#Compute the Critical Poset
critical_poset(X, M)
```

```python
#Compute the localizaton poset
localization_poset(X, M)
```

### Example 2

```python
# The boundary of a 2-simplex
K = simplicial_complexes.Sphere(2)
X = K.face_poset()
X.show()
```

```python
# Acyclic matching
M = [((3,),(2,3)), ((1,),(0,1)), ((0,2), (0,1,2)), ((0,3), (0,2,3))]
```

```python
#Flow Category
Flow = flow_category(X, M)
```

```python
Crit = critical_poset(X, M)
Crit.plot()
```

```python
# Check contractibility of Flow Posets

C = critical_points(X, M)

for x in C:
    for y in C:
        if len(Flow[(x, y)])>0:
            print(x,y, is_homotopically_trivial(Flow[(x, y)]))
            
```

```python
L = localization_poset(X, M)
L.show()
```
